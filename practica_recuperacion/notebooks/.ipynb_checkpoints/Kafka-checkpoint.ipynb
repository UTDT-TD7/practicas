```python
from kafka import KafkaConsumer
```

```python
consumer = KafkaConsumer("postgres.public.ejemplo",bootstrap_servers="broker:29092")
```

```python
consumer.topics()
```

```python
from rich import print
import json

for msg in consumer:
    print(msg)
    if msg.key is not None:
        print("key:")
        print(json.loads(msg.key.decode()))
    if msg.value is not None:
        print("value:")
        print(json.loads(msg.value.decode()))
    print("\n")
```
